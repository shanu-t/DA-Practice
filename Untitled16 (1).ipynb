```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset("iris")
sns.lineplot(x = "sepal_length", y = "sepal_width", data=data)
sns.set_style("darkgrid")
plt.title("Framing by Sepal Length and Width")
plt.xlim()
plt.show()
```

```python
import numpy as np
import seaborn as sns
data = sns.load_dataset("iris")
def makeplot():
    sns.lineplot(x = "sepal_length", y = "sepal_width", data= data)
with sns.axes_style("white"):
    plt.subplot(2,1,1)
    makeplot()
plt.subplot(2,1,2)
makeplot()
```

```python
import numpy as np
import seaborn as sns
data = sns.load_dataset("iris")
def makeplot():
    sns.lineplot(x = "sepal_length", y = "sepal_width", data= data)
sns.set_style("white") 
plt.subplot(211)
sns.set_palette("Accent")
makeplot()
plt.subplot(212)
sns.set_palette("vlag")
makeplot()
plt.show()
```
